```python
# ensure Annotation

def get_prodict(x:int, y:int) -> int:
    return x*y
```

```python
get_prodict(2,3)
```

```python
from ensure import ensure_annotations
```

```python
@ensure_annotations
def get_prodict(x:int, y:int) -> int:
    return x*y
```

```python
get_prodict(2,3)
```
